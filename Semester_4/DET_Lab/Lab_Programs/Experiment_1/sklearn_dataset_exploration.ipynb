```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df["target"] = cancer.target
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
df
```

```python
print(df.isna().sum())
```

```python
df.describe()
```

```python
numeric = df.select_dtypes(include="number")
column_means = numeric.mean()
column_medians = numeric.median()
column_modes = numeric.mode().iloc[0]
column_std = numeric.std()

print("Means:")
print(column_means.head(10))
print("Medians:")
print(column_medians.head(10))
print("Modes:")
print(column_modes.head(10))
print("Standard deviation:")
print(column_std.head(10))
```

```python
import matplotlib.pyplot as plt

stats_df = pd.DataFrame({
    "Mean": column_means,
    "Median": column_medians,
    "Mode": column_modes,
    "Std Dev": column_std
})

stats_df.plot(kind="bar", figsize=(15,6))

plt.title("Statistical Summary of Breast Cancer Dataset")
plt.xlabel("Features")
plt.ylabel("Value")
plt.legend()
plt.tight_layout()
plt.show()
```
