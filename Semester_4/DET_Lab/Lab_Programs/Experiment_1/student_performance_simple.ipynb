```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/sujal/OneDrive/Desktop/Study/Btech/Sem 4/DETL/Exp 1/simple_student_data_v2.csv")
```

```python
score_cols = ['math', 'science', 'english', 'history', 'sports', 'attendance']
missing_counts = df.isna().sum()
print('Missing values per column:')
print(missing_counts)
```

```python
df_filled = df.copy()
df_filled[score_cols] = df_filled[score_cols].fillna(df_filled[score_cols].mean())
```

```python
df_filled['total'] = df_filled[score_cols].sum(axis=1)
df_filled['average'] = df_filled[score_cols].mean(axis=1)
print(df_filled[['name'] + score_cols + ['total', 'average']].head())
```

```python
# Basic stats with built-in functions
column_means = df_filled[score_cols].mean()
column_medians = df_filled[score_cols].median()
column_modes = df_filled[score_cols].mode().iloc[0]
column_std = df_filled[score_cols].std()

print('Means:')
print(column_means)
print('Medians:')
print(column_medians)
print('Modes:')
print(column_modes)
print('Standard deviation:')
print(column_std)
```
