## Task 4: Identify and Remove Duplicate Records
### Multiple Methods to Handle Duplicates

```python
print("=" * 50)
print("DUPLICATE DETECTION")
print("=" * 50)

# Method 1: Check for any duplicates
print("\n1. Are there any duplicate rows?")
print(df_academic.duplicated().any())

# Method 2: Count duplicate rows
print("\n2. Number of duplicate rows:")
print(df_academic.duplicated().sum())

# Method 3: Show duplicate rows
print("\n3. Display duplicate rows:")
duplicates = df_academic[df_academic.duplicated(keep=False)]
print(duplicates)

# Method 4: Check duplicates based on specific column (Student_ID)
print("\n4. Duplicates based on Student_ID:")
id_duplicates = df_academic[df_academic.duplicated(subset=['Student_ID'], keep=False)]
print(id_duplicates.sort_values('Student_ID'))

# Method 5: Count duplicates per Student_ID
print("\n5. Count of each Student_ID:")
id_counts = df_academic['Student_ID'].value_counts()
print(id_counts[id_counts > 1])
```

```python
print("=" * 50)
print("REMOVING DUPLICATES")
print("=" * 50)

print(f"\nOriginal shape: {df_academic.shape}")

# Method 1: Remove exact duplicate rows (keep first occurrence)
df_no_dup1 = df_academic.drop_duplicates(keep='first')
print(f"\n1. After removing exact duplicates (keep='first'): {df_no_dup1.shape}")

# Method 2: Remove duplicates based on Student_ID (keep first)
df_no_dup2 = df_academic.drop_duplicates(subset=['Student_ID'], keep='first')
print(f"\n2. After removing Student_ID duplicates (keep='first'): {df_no_dup2.shape}")

# Method 3: Keep last occurrence
df_no_dup3 = df_academic.drop_duplicates(subset=['Student_ID'], keep='last')
print(f"\n3. After removing Student_ID duplicates (keep='last'): {df_no_dup3.shape}")

# Method 4: Remove all duplicates (don't keep any)
df_no_dup4 = df_academic.drop_duplicates(subset=['Student_ID'], keep=False)
print(f"\n4. After removing all Student_ID duplicates (keep=False): {df_no_dup4.shape}")

# We'll use Method 2 for further processing
df_academic_clean = df_academic.drop_duplicates(subset=['Student_ID'], keep='first')
print(f"\n✓ Final cleaned data shape: {df_academic_clean.shape}")
print("\nCleaned data (first 10 rows):")
print(df_academic_clean.head(10))
```

## Task 5: Detect and Handle Missing Values
### Multiple Techniques for Handling Missing Data

```python
print("=" * 50)
print("MISSING VALUE DETECTION")
print("=" * 50)

# Method 1: Count missing values per column
print("\n1. Missing values per column:")
missing_count = df_academic_clean.isnull().sum()
print(missing_count)

# Method 2: Percentage of missing values
print("\n2. Percentage of missing values:")
missing_percent = (df_academic_clean.isnull().sum() / len(df_academic_clean)) * 100
print(missing_percent)

# Method 3: Total missing values
print("\n3. Total missing values in dataset:")
print(df_academic_clean.isnull().sum().sum())

# Method 4: Rows with any missing value
print("\n4. Number of rows with at least one missing value:")
print(df_academic_clean.isnull().any(axis=1).sum())

# Method 5: Display rows with missing values
print("\n5. Rows containing missing values:")
rows_with_missing = df_academic_clean[df_academic_clean.isnull().any(axis=1)]
print(rows_with_missing)

# Visualize missing data
print("\n6. Missing Data Summary:")
missing_df = pd.DataFrame({
    'Column': df_academic_clean.columns,
    'Missing_Count': df_academic_clean.isnull().sum().values,
    'Missing_Percentage': missing_percent.values
})
print(missing_df[missing_df['Missing_Count'] > 0])
```

```python
print("=" * 50)
print("HANDLING MISSING VALUES - DIFFERENT METHODS")
print("=" * 50)

# Get only numeric columns for imputation
numeric_cols = ['Math', 'Science', 'English', 'History', 'Geography']

# Method 1: Fill with Mean
print("\n1. MEAN IMPUTATION")
df_mean = df_academic_clean.copy()
for col in numeric_cols:
    mean_value = df_mean[col].mean()
    df_mean[col].fillna(mean_value, inplace=True)
    print(f"{col} mean: {mean_value:.2f}")
print(f"Missing values after mean imputation: {df_mean.isnull().sum().sum()}")

# Method 2: Fill with Median
print("\n2. MEDIAN IMPUTATION")
df_median = df_academic_clean.copy()
for col in numeric_cols:
    median_value = df_median[col].median()
    df_median[col].fillna(median_value, inplace=True)
    print(f"{col} median: {median_value:.2f}")
print(f"Missing values after median imputation: {df_median.isnull().sum().sum()}")

# Method 3: Fill with Mode
print("\n3. MODE IMPUTATION")
df_mode = df_academic_clean.copy()
for col in numeric_cols:
    mode_value = df_mode[col].mode()[0]  # mode() returns a Series, take first value
    df_mode[col].fillna(mode_value, inplace=True)
    print(f"{col} mode: {mode_value:.2f}")
print(f"Missing values after mode imputation: {df_mode.isnull().sum().sum()}")

# Method 4: Fill with a specific value
print("\n4. CONSTANT VALUE IMPUTATION")
df_constant = df_academic_clean.copy()
df_constant[numeric_cols] = df_constant[numeric_cols].fillna(0)
print(f"Missing values after constant (0) imputation: {df_constant.isnull().sum().sum()}")

# Method 5: Forward Fill (use previous value)
print("\n5. FORWARD FILL")
df_ffill = df_academic_clean.copy()
df_ffill[numeric_cols] = df_ffill[numeric_cols].fillna(method='ffill')
print(f"Missing values after forward fill: {df_ffill.isnull().sum().sum()}")

# Method 6: Backward Fill (use next value)
print("\n6. BACKWARD FILL")
df_bfill = df_academic_clean.copy()
df_bfill[numeric_cols] = df_bfill[numeric_cols].fillna(method='bfill')
print(f"Missing values after backward fill: {df_bfill.isnull().sum().sum()}")

# Method 7: Interpolation (linear)
print("\n7. LINEAR INTERPOLATION")
df_interpolate = df_academic_clean.copy()
df_interpolate[numeric_cols] = df_interpolate[numeric_cols].interpolate(method='linear')
print(f"Missing values after interpolation: {df_interpolate.isnull().sum().sum()}")

# Method 8: Drop rows with missing values
print("\n8. DROP ROWS WITH MISSING VALUES")
df_dropna = df_academic_clean.copy()
df_dropna_rows = df_dropna.dropna()
print(f"Shape before: {df_dropna.shape}")
print(f"Shape after dropping rows: {df_dropna_rows.shape}")

# Method 9: Drop columns with missing values
print("\n9. DROP COLUMNS WITH MISSING VALUES")
df_dropna_cols = df_dropna.dropna(axis=1)
print(f"Columns before: {df_dropna.shape[1]}")
print(f"Columns after: {df_dropna_cols.shape[1]}")

# For further analysis, we'll use MEAN IMPUTATION
print("\n" + "="*50)
print("✓ Using MEAN IMPUTATION for further processing")
print("="*50)
df_academic_clean = df_mean.copy()
```

## Task 6: Integrate Academic and Attendance Data
### Different Types of Merging/Joining

```python
print("=" * 50)
print("DATA INTEGRATION - DIFFERENT MERGE TYPES")
print("=" * 50)

print(f"\nAcademic data shape: {df_academic_clean.shape}")
print(f"Attendance data shape: {df_attendance.shape}")

# Method 1: Inner Join (only matching records)
print("\n1. INNER JOIN (only matching Student_IDs):")
df_inner = pd.merge(df_academic_clean, df_attendance, on='Student_ID', how='inner')
print(f"Shape: {df_inner.shape}")
print(df_inner.head())

# Method 2: Left Join (all academic records)
print("\n2. LEFT JOIN (all academic records):")
df_left = pd.merge(df_academic_clean, df_attendance, on='Student_ID', how='left')
print(f"Shape: {df_left.shape}")

# Method 3: Right Join (all attendance records)
print("\n3. RIGHT JOIN (all attendance records):")
df_right = pd.merge(df_academic_clean, df_attendance, on='Student_ID', how='right')
print(f"Shape: {df_right.shape}")

# Method 4: Outer Join (all records)
print("\n4. OUTER JOIN (all records from both):")
df_outer = pd.merge(df_academic_clean, df_attendance, on='Student_ID', how='outer')
print(f"Shape: {df_outer.shape}")

# Method 5: Using concat (stacking)
print("\n5. CONCAT (for demonstration - not recommended for this case):")
# This is just to show concat - not useful for this specific case
df_concat = pd.concat([df_academic_clean, df_attendance], axis=1)
print(f"Shape: {df_concat.shape}")

# Use inner join for final dataset
df_final = df_inner.copy()
print("\n" + "="*50)
print("✓ Using INNER JOIN for final dataset")
print(f"✓ Final merged dataset shape: {df_final.shape}")
print("="*50)
print("\nFinal merged data (first 5 rows):")
print(df_final.head())
```

## Task 8: Normalize Numerical Attributes
### Min-Max and Z-Score Normalization

```python
print("=" * 50)
print("DATA NORMALIZATION")
print("=" * 50)

# Select columns to normalize
cols_to_normalize = ['Math', 'Science', 'English', 'History', 'Geography']

print("\nOriginal data (first 5 rows):")
print(df_final[cols_to_normalize].head())
print("\nOriginal Statistics:")
print(df_final[cols_to_normalize].describe())
```

```python
# Method 1: Min-Max Normalization (scales to 0-1)
print("\n" + "="*50)
print("METHOD 1: MIN-MAX NORMALIZATION (0 to 1)")
print("="*50)

# Using sklearn
scaler_minmax = MinMaxScaler()
df_minmax = df_final.copy()
df_minmax[cols_to_normalize] = scaler_minmax.fit_transform(df_final[cols_to_normalize])

print("\nAfter Min-Max Normalization:")
print(df_minmax[cols_to_normalize].head())
print("\nStatistics after Min-Max:")
print(df_minmax[cols_to_normalize].describe())

# Manual Min-Max Normalization
print("\n--- Manual Min-Max Formula ---")
print("normalized = (value - min) / (max - min)")
df_minmax_manual = df_final.copy()
for col in cols_to_normalize:
    min_val = df_final[col].min()
    max_val = df_final[col].max()
    df_minmax_manual[col] = (df_final[col] - min_val) / (max_val - min_val)
    print(f"{col}: min={min_val:.2f}, max={max_val:.2f}")

print("\nManual Min-Max result (first 5):")
print(df_minmax_manual[cols_to_normalize].head())
```

```python
# Method 2: Z-Score Normalization (Standardization)
print("\n" + "="*50)
print("METHOD 2: Z-SCORE NORMALIZATION (Standardization)")
print("="*50)

# Using sklearn
scaler_standard = StandardScaler()
df_zscore = df_final.copy()
df_zscore[cols_to_normalize] = scaler_standard.fit_transform(df_final[cols_to_normalize])

print("\nAfter Z-Score Normalization:")
print(df_zscore[cols_to_normalize].head())
print("\nStatistics after Z-Score:")
print(df_zscore[cols_to_normalize].describe())

# Manual Z-Score Normalization
print("\n--- Manual Z-Score Formula ---")
print("z-score = (value - mean) / std_deviation")
df_zscore_manual = df_final.copy()
for col in cols_to_normalize:
    mean_val = df_final[col].mean()
    std_val = df_final[col].std()
    df_zscore_manual[col] = (df_final[col] - mean_val) / std_val
    print(f"{col}: mean={mean_val:.2f}, std={std_val:.2f}")

print("\nManual Z-Score result (first 5):")
print(df_zscore_manual[cols_to_normalize].head())
```

```python
# Method 3: Custom Range Normalization (e.g., 0-100)
print("\n" + "="*50)
print("METHOD 3: CUSTOM RANGE NORMALIZATION (0 to 100)")
print("="*50)

df_custom = df_final.copy()
for col in cols_to_normalize:
    min_val = df_final[col].min()
    max_val = df_final[col].max()
    # Scale to 0-100
    df_custom[col] = ((df_final[col] - min_val) / (max_val - min_val)) * 100

print("\nAfter Custom Range (0-100) Normalization:")
print(df_custom[cols_to_normalize].head())
print("\nStatistics after Custom Range:")
print(df_custom[cols_to_normalize].describe())
```

```python
# Method 4: Robust Scaling (using median and IQR)
print("\n" + "="*50)
print("METHOD 4: ROBUST SCALING (using median and IQR)")
print("="*50)
print("Formula: (value - median) / IQR")

df_robust = df_final.copy()
for col in cols_to_normalize:
    median_val = df_final[col].median()
    q1 = df_final[col].quantile(0.25)
    q3 = df_final[col].quantile(0.75)
    iqr = q3 - q1
    df_robust[col] = (df_final[col] - median_val) / iqr
    print(f"{col}: median={median_val:.2f}, IQR={iqr:.2f}")

print("\nAfter Robust Scaling:")
print(df_robust[cols_to_normalize].head())
```

```python
# Comparison of normalization methods
print("\n" + "="*50)
print("COMPARISON OF NORMALIZATION METHODS")
print("="*50)

# Take first student's Math score as example
original_value = df_final.iloc[0]['Math']
print(f"\nOriginal Math score (first student): {original_value}")
print(f"Min-Max normalized: {df_minmax.iloc[0]['Math']:.4f}")
print(f"Z-Score normalized: {df_zscore.iloc[0]['Math']:.4f}")
print(f"Custom Range (0-100): {df_custom.iloc[0]['Math']:.4f}")
print(f"Robust scaled: {df_robust.iloc[0]['Math']:.4f}")

# Visualize different normalizations
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

axes[0, 0].hist(df_final['Math'], bins=15, edgecolor='black')
axes[0, 0].set_title('Original Data')
axes[0, 0].set_xlabel('Math Scores')

axes[0, 1].hist(df_minmax['Math'], bins=15, edgecolor='black')
axes[0, 1].set_title('Min-Max Normalized')
axes[0, 1].set_xlabel('Normalized Scores')

axes[0, 2].hist(df_zscore['Math'], bins=15, edgecolor='black')
axes[0, 2].set_title('Z-Score Normalized')
axes[0, 2].set_xlabel('Z-Scores')

axes[1, 0].hist(df_custom['Math'], bins=15, edgecolor='black')
axes[1, 0].set_title('Custom Range (0-100)')
axes[1, 0].set_xlabel('Scaled Scores')

axes[1, 1].hist(df_robust['Math'], bins=15, edgecolor='black')
axes[1, 1].set_title('Robust Scaled')
axes[1, 1].set_xlabel('Robust Scores')

axes[1, 2].axis('off')

plt.tight_layout()
plt.savefig('normalization_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
print("\n✓ Normalization comparison saved as 'normalization_comparison.png'")
```

## Task 9: Reduce Dataset by Removing Irrelevant/Redundant Attributes
### Feature Selection and Dimensionality Reduction

```python
print("=" * 50)
print("FEATURE REDUCTION")
print("=" * 50)

print(f"\nOriginal dataset shape: {df_final.shape}")
print(f"Columns: {df_final.columns.tolist()}")

# Method 1: Remove columns with high correlation (redundant)
print("\n1. Identifying Highly Correlated Features:")
correlation_matrix = df_final[subject_cols].corr().abs()
upper_triangle = correlation_matrix.where(
    np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool)
)

# Find features with correlation > 0.9
high_corr_features = [column for column in upper_triangle.columns 
                      if any(upper_triangle[column] > 0.9)]
print(f"Highly correlated features (>0.9): {high_corr_features}")
if not high_corr_features:
    print("No highly correlated features found (threshold: 0.9)")

# Method 2: Remove low variance features
print("\n2. Variance Analysis:")
variances = df_final[subject_cols].var()
print("Variance of each subject:")
print(variances)
low_var_threshold = 10  # You can adjust this
low_variance_features = variances[variances < low_var_threshold].index.tolist()
print(f"\nLow variance features (<{low_var_threshold}): {low_variance_features}")

# Method 3: Remove based on missing value percentage
print("\n3. Missing Value Percentage (before imputation check):")
# Note: We already handled missing values, but showing the method
missing_pct = (df_final.isnull().sum() / len(df_final)) * 100
print(missing_pct)
high_missing_cols = missing_pct[missing_pct > 30].index.tolist()
print(f"Columns with >30% missing: {high_missing_cols}")
```

```python
# Method 4: Select most important features based on domain knowledge
print("\n4. Feature Selection Based on Importance:")

# Calculate which subjects contribute most to overall performance
df_final['Overall_Score'] = df_final[subject_cols].mean(axis=1)

importance_scores = {}
for col in subject_cols:
    correlation = df_final[col].corr(df_final['Overall_Score'])
    importance_scores[col] = correlation

importance_df = pd.DataFrame(list(importance_scores.items()), 
                            columns=['Subject', 'Importance'])
importance_df = importance_df.sort_values('Importance', ascending=False)
print("\nFeature Importance (correlation with overall score):")
print(importance_df)

# Visualize importance
plt.figure(figsize=(10, 6))
plt.barh(importance_df['Subject'], importance_df['Importance'])
plt.xlabel('Correlation with Overall Score')
plt.title('Feature Importance')
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=300, bbox_inches='tight')
plt.show()
print("\n✓ Feature importance chart saved")
```

```python
# Create reduced dataset
print("\n" + "="*50)
print("CREATING REDUCED DATASET")
print("="*50)

# Keep only essential columns
# Removing: Days_Present and Days_Absent (redundant with Attendance_Percentage)
# Removing: Geography (if needed, based on importance)
# Removing: computed columns like Total_Marks, Highest_Subject_Score, etc.

essential_columns = [
    'Student_ID',
    'Name',
    'Math',
    'Science',
    'English',
    'History',
    'Attendance_Percentage'
]

df_reduced = df_final[essential_columns].copy()

print(f"\nOriginal dataset shape: {df_final.shape}")
print(f"Reduced dataset shape: {df_reduced.shape}")
print(f"Columns removed: {set(df_final.columns) - set(essential_columns)}")
print(f"\nReduced dataset columns: {df_reduced.columns.tolist()}")
print("\nReduced dataset (first 10 rows):")
print(df_reduced.head(10))
```

```python
# Alternative: PCA (Principal Component Analysis) for dimensionality reduction
print("\n" + "="*50)
print("BONUS: PCA DIMENSIONALITY REDUCTION")
print("="*50)

from sklearn.decomposition import PCA

# Apply PCA on subject scores
pca = PCA(n_components=3)  # Reduce 5 subjects to 3 components
subject_scores = df_final[subject_cols]
pca_result = pca.fit_transform(subject_scores)

# Create DataFrame with PCA components
df_pca = pd.DataFrame(data=pca_result, 
                      columns=['PC1', 'PC2', 'PC3'])
df_pca['Student_ID'] = df_final['Student_ID'].values
df_pca['Name'] = df_final['Name'].values

print("\nExplained variance ratio:")
print(pca.explained_variance_ratio_)
print(f"\nTotal variance explained: {sum(pca.explained_variance_ratio_):.2%}")

print("\nPCA transformed data (first 5 rows):")
print(df_pca.head())

# Visualize PCA
plt.figure(figsize=(8, 6))
plt.bar(range(1, 4), pca.explained_variance_ratio_)
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.title('PCA - Variance Explained by Each Component')
plt.xticks(range(1, 4))
plt.tight_layout()
plt.savefig('pca_variance.png', dpi=300, bbox_inches='tight')
plt.show()
print("\n✓ PCA variance chart saved")
```
