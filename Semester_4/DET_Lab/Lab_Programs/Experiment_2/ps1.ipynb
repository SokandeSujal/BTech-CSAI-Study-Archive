## Task 7: Compute basic statistical measures such as mean, median, and standard deviation

```python
subject_cols = ['Math', 'Science', 'English', 'History', 'Geography']

print("\n1. MEAN (Average):")
means = df_final[subject_cols].mean()
print(means)

print("\n2. MEDIAN (Middle value):")
medians = df_final[subject_cols].median()
print(medians)

print("\n3. MODE (Most frequent):")
modes = df_final[subject_cols].mode().iloc[0]
print(modes)

print("\n4. STANDARD DEVIATION:")
std_devs = df_final[subject_cols].std()
print(std_devs)

print("\n5. VARIANCE:")
variances = df_final[subject_cols].var()
print(variances)

print("\n6. MINIMUM VALUES:")
minimums = df_final[subject_cols].min()
print(minimums)

print("\n7. MAXIMUM VALUES:")
maximums = df_final[subject_cols].max()
print(maximums)

print("\n8. RANGE (Max - Min):")
ranges = maximums - minimums
print(ranges)
```

```python
import seaborn as sns

print("\nCorrelation Matrix:")
correlation_matrix = df_final[subject_cols].corr()
print(correlation_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, fmt='.2f')
plt.title('Correlation Matrix of Subjects')
plt.tight_layout()
plt.savefig('correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()
print("\n✓ Correlation matrix saved as 'correlation_matrix.png'")
```

## Task 9: Reduce Dataset by Removing Irrelevant/Redundant Attributes
### Feature Selection and Dimensionality Reduction

```python
correlation_matrix = df_final[subject_cols].corr().abs()
```

```python
correlation_matrix
```

```python
# Create reduced dataset
print("\n" + "="*50)
print("CREATING REDUCED DATASET")
print("="*50)

# Keep only essential columns
# Removing: Days_Present and Days_Absent (redundant with Attendance_Percentage)
# Removing: Geography (if needed, based on importance)
# Removing: computed columns like Total_Marks, Highest_Subject_Score, etc.

essential_columns = [
    'Student_ID',
    'Name',
    'Math',
    'Science',
    'English',
    'History',
    'Attendance_Percentage'
]

df_reduced = df_final[essential_columns].copy()

print(f"\nOriginal dataset shape: {df_final.shape}")
print(f"Reduced dataset shape: {df_reduced.shape}")
print(f"Columns removed: {set(df_final.columns) - set(essential_columns)}")
print(f"\nReduced dataset columns: {df_reduced.columns.tolist()}")
print("\nReduced dataset (first 10 rows):")
print(df_reduced.head(10))
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
subject_scores = df_final[subject_cols]
pca_result = pca.fit_transform(subject_scores)

df_pca = pd.DataFrame(data=pca_result, 
                      columns=['PC1', 'PC2', 'PC3'])
df_pca['Student_ID'] = df_final['Student_ID'].values
df_pca['Name'] = df_final['Name'].values

plt.figure(figsize=(8, 6))
plt.bar(range(1, 4), pca.explained_variance_ratio_)
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.title('PCA - Variance Explained by Each Component')
plt.xticks(range(1, 4))
plt.tight_layout()
plt.savefig('pca_variance.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Task 10: Validate and Export Final Preprocessed Dataset

```python
print("\n1. Check for missing values:")
missing_check = df_reduced.isnull().sum()
print(missing_check)
if missing_check.sum() == 0:
    print("✓ No missing values found!")
else:
    print("✗ Missing values detected!")

print("\n2. Check for duplicates:")
duplicate_check = df_reduced.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_check}")
if duplicate_check == 0:
    print("✓ No duplicates found!")
else:
    print("✗ Duplicates detected!")

print("\n3. Check data types:")
print(df_reduced.dtypes)

print("\n4. Check for negative values (shouldn't exist in marks):")
numeric_cols = ['Math', 'Science', 'English', 'History', 'Attendance_Percentage']
negative_check = (df_reduced[numeric_cols] < 0).any()
print(negative_check)
if not negative_check.any():
    print("✓ No negative values found!")
else:
    print("✗ Negative values detected!")

print("\n5. Check value ranges:")
print("Should be between 0-100 for marks and 0-100 for attendance")
for col in numeric_cols:
    min_val = df_reduced[col].min()
    max_val = df_reduced[col].max()
    print(f"{col}: min={min_val:.2f}, max={max_val:.2f}")
    if 0 <= min_val and max_val <= 100:
        print(f"  ✓ {col} is in valid range")
    else:
        print(f"  ✗ {col} has out-of-range values")

print("\n6. Statistical summary of final data:")
print(df_reduced[numeric_cols].describe())

print("\n7. Check for unique Student IDs:")
unique_ids = df_reduced['Student_ID'].nunique()
total_rows = len(df_reduced)
print(f"Unique Student IDs: {unique_ids}")
print(f"Total rows: {total_rows}")
if unique_ids == total_rows:
    print("✓ All Student IDs are unique!")
else:
    print("✗ Duplicate Student IDs found!")
```
